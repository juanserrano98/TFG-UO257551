```python
import numpy as np
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt


m = np.loadtxt("C:/Users/xuans/OneDrive/Escritorio/CEMENTERA/ficheros_motores/Motor_1/Motor1_AxeY_full/AxeY/1667738780-5000Hz-DE.csv") 

F = 5000 # Sampling frequency
N = 32768 # Number of samples

timestep = 1/F
window = np.hanning(N)
fft_res = fft(m*window)
fft_res = np.abs(fft_res[0:int(N/2)])/N*2

fft_freq = fftfreq(32768,d=timestep)
fft_freq = fft_freq[0:int(N/2)]

plt.plot(fft_freq,fft_res)
"""marker = '.'"""

#pintado de secciones de cada armonico especial
#busqueda para cada armonico y representacion para ver si hay algo interesante
#1X: 
#BPFO: si parto de la pos 110, +-2Hz seria de 2/0,15 = 14 posiciones arriba/abajo de 96 a 124
x = 952
#desviacion 
y = 50
z = x - y
z2 = x + y
max_value = np.argmax(fft_res[z:z2])
max_value = z + max_value
print(max_value)

plt.xlim(fft_freq[z],fft_freq[z2])
plt.title("Fallo BSF, muestra Ãºnica EJE Y")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
```
